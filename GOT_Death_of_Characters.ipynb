# **2. DATA CLEANING**

```python
# Create a new death dataset and select only the required columns from the original death dataset, which we import in the jupyter notebook.
death=death[['Name','Allegiances','Death Year','Book of Death','Death Chapter','Gender','Nobility']]
death
```

```python
# Number of rows and columns in the dataset.
death.shape
```

```python
# Datatypes of all the columns of the dataset.
death.dtypes
```

```python
# Check all the null values in dataset.
death.isnull()
```

```python
# Sum all the null values in the dataset.
death.isnull().sum()
```

```python
# To avoid settingWithCopyWarning
pd.options.mode.chained_assignment=None
```

```python
# Change the data type of the Gender column to boolean.
death['Gender'] = death['Gender'].astype('bool')
```

```python
# Change the data type of the Nobality column to boolean.
death['Nobility'] = death['Nobility'].astype('bool')
```

```python
# Averge of all the values of Death Year Column.
(299+300+298+297)/4
```

```python
# Replace all null values in the Death Year column with 298.5.
death.fillna({'Death Year':298.5},inplace=True)
```

```python
# Convert datatype of death Year column into integer.
death['Death Year'] = death['Death Year'].fillna(0).astype(int)
```

```python
# Averge of all the values of Book of Death Column.
(3+2+5+1+4)/4
```

```python
# Replace all null values in the Book of Death column with 3.75.
death.fillna({'Book of Death':3.75},inplace=True)
```

```python
# Convert datatype of Book of Death column into integer.
death['Book of Death'] = death['Book of Death'].fillna(0).astype(int)
```

```python
# Mean of all the values od Death Chapter Column.
death['Death Chapter'].mean()
```

```python
# Replace all null values in the Death Chapter column with 40.070234113712374.
death.fillna({'Death Chapter':40.070234113712374},inplace=True)
```

```python
# Convert datatype of Death Chapter column into integer.
death['Death Chapter'] = death['Death Chapter'].fillna(0).astype(int)
```

```python
# Datatypes of all the columns in the dataset.
death.dtypes
```

```python
# Sum of all null values in the dataset.
death.isnull().sum()
```
